# Graph comparison in python with Matplotlib

Plotting graph of LeNet model 

```python
import matplotlib.pyplot as plt 
x = [10,20,30,40,50,60] 
# corresponding y axis values 
y = [48.625001311302185,57.749998569488525,82.12500214576721,99.87499713897705,100.0,100.0] 
  
# plotting the points  
plt.plot(x, y) 
  
# naming the x axis 
plt.xlabel('percentage noise') 
# naming the y axis 
plt.ylabel('accuracy') 
  
# giving a title to my graph 
plt.title('LeNet Model') 
  
# function to show the plot 
plt.show() 
```

Plotting graph of ResNet Model

```python
x = [10,20,30,40,50,60] 
# corresponding y axis values 
y = [74.84375238418579,86.93749904632568,99.40624833106995,99.75000023841858,99.97222423553467,98.06249737739563] 
  
# plotting the points  
plt.plot(x, y) 
  
# naming the x axis 
plt.xlabel('percentage noise') 
# naming the y axis 
plt.ylabel('accuracy') 
  
# giving a title to my graph 
plt.title('ResNet Model') 
  
# function to show the plot 
plt.show() 
```

Comparing Both the graphs

```python
import matplotlib.pyplot as plt 

# line 1 points 
x1 = [10,20,30,40,50,60] 
y1 = [48.625001311302185,57.749998569488525,82.12500214576721,99.87499713897705,100.0,100.0]
# plotting the line 1 points 
plt.plot(x1, y1, label = "LeNet Model") 

# line 2 points 
x2 = [10,20,30,40,50,60]  
y2 = [74.84375238418579,86.93749904632568,99.40624833106995,99.75000023841858,99.97222423553467,98.06249737739563] 
# plotting the line 2 points 
plt.plot(x2, y2, label = "ResNet Model") 

# naming the x axis 
plt.xlabel('percentage noise') 
# naming the y axis 
plt.ylabel('accuracy') 
# giving a title to my graph 
plt.title('accuracy vs percentage noise comparison') 

# show a legend on the plot 
plt.legend() 

# function to show the plot 
plt.show() 
```
